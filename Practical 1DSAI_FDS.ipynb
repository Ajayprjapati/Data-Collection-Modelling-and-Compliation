```python
import pandas as pd
names = ['Bob', 'Rio', 'Joy', 'Larry', 'Riks']
grades = [76, 95, 65, 66, 45]
GradeList = zip(names, grades)
df = pd.DataFrame(data = GradeList,columns = ['Names', 'Grades'])

writer = pd.ExcelWriter('dataframe.xlsx', engine='xlswriter')
df.to_excel(writer, sheet_name='sheet1')
writer.save()
```
